```python
with open ("input.txt") as fin:
  grid=fin.read().strip().split("\n")

from collections import defaultdict
from itertools import combinations

n=len(grid)

def in_bound(x,y):
  return 0 <= x < n  and 0 <= y <n


def get_antinodes(a,b):
  ax,ay=a
  bx,by=b

  dx,dy=bx-ax,by-ay

  i=0

  while True:
    if in_bound(ax - dx * i, ay - dy * i):
      yield (ax - dx * i, ay - dy * i)
    else:
      break
    i+=1

  i=0
  while True:
    if in_bound(bx + dx * i, by + dy * i):
      yield (bx + dx * i, by + dy * i)
    else:
      break
    i+=1
antinodes=set()

all_locs = defaultdict(list)

for i in range(n):
  for j in range(n):
    if grid[i][j] != ".":
      all_locs[grid[i][j]].append((i,j))


for freq in all_locs:
  loc=all_locs[freq]
  for a,b in combinations(loc,r=2):
    for antinode in get_antinodes(a,b):
      antinodes.add(antinode)

print(len(antinodes))
```
