```python
with open("input.txt") as fin:
  grid=fin.read().strip().split("\n")

n=len(grid)

dd=[[0,1],[1,0],[0,-1],[-1,0]]

def in_grid(i,j):
  return (0<=i<n)and (0<=j<n)

def score(i,j):
  if grid[i][j] != "0":
    return 0

  ans=0 # Initialize ans inside the function
  stack=[(i,j)]
  visited=set()
  while len(stack)>0:
    curi,curj=stack.pop()
    cur=grid[curi][curj]
    if (curi,curj) in visited: # Check if tuple is in visited
      continue
    visited.add((curi,curj))

    if cur == '9' : # Compare cur with string '9'
      ans +=1
      continue

    for di,dj in dd:
      ni,nj=curi+di,curj+dj
      if not in_grid(ni,nj): # Pass ni, nj to in_grid
        continue
      nbr=grid[ni][nj]
      if nbr != str(int(cur) + 1): # Convert cur and nbr to int for comparison
        continue
      stack.append((ni,nj)) # Append ni, nj to stack
  return ans

ans =0
for i in range(n):
  for j in range(n): # Add a nested loop for j
    ans+=score(i,j)

print(ans)
```

```python
from functools import lru_cache
with open("input.txt") as fin:
  grid=fin.read().strip().split("\n")

n=len(grid)

dd=[[0,1],[1,0],[0,-1],[-1,0]]
def in_grid(i,j):
  return (0<=i<n)and (0<=j<n)

@lru_cache(None)

def rating(i,j):
  if grid[i][j] == "9":
    return 1
  ans=0
  for di, dj in dd:
        ii, jj = i + di, j + dj
        if not in_grid(ii, jj):
            continue

        if int(grid[ii][jj]) == int(grid[i][j]) + 1:
            ans += rating(ii, jj)

  return ans


ans = 0
for i in range(n):
    for j in range(n):
        if grid[i][j] == "0":
            ans += rating(i, j)

print(ans)
```
