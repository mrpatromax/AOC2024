```python
total_sum = 0
mul_enable = True

def multiply(memory):
  pattern = re.compile(r"mul\((\d{1,3}),(\d{1,3})\)")
  do_patt = re.compile(r"do\(\)")  # Changed to "do()"
  dont_patt = re.compile(r"don't\(\)")  # Changed to "don't()"

  total_sum = 0
  mul_enable = True

  tokens = re.split(r"(do\(\)|don't\(\)|mul\(\d{1,3},\d{1,3}\))", memory)
  for token in tokens:  # Changed 'tokens' to 'token' for clarity
    if not token:
      continue
    if do_patt.fullmatch(token):  # Pass 'token' to fullmatch
      mul_enable = True
    elif dont_patt.fullmatch(token):  # Pass 'token' to fullmatch
      mul_enable = False
    elif mul_enable and pattern.fullmatch(token):  # Pass 'token' to fullmatch
      x, y = map(int, pattern.match(token).groups())  # Changed to groups() and fixed typo
      total_sum += x * y  # Fixed typo 'total_sun' to 'total_sum'
  return total_sum

if __name__ == "__main__":
  with open("input.txt") as file:
    croupted_memory = file.read()
    print(multiply(croupted_memory))
```
