```python
with open("input4.txt") as fin:
    # Split the line into individual numbers and convert them to integers
    stones = [int(x) for x in fin.read().strip().split()]
from functools import cache

@cache
def count(stone, steps):
    if steps == 0:
        return 1
    if stone == 0:  # Changed stones to stone
        return count(1, steps - 1)

    string = str(stone) # Changed stones to stone
    length = len(string)
    if length % 2 == 0:
        return count(int(string[0:length // 2]), steps - 1) + count(int(string[length // 2:]), steps - 1)
    return count(stone * 2024, steps - 1)

print(sum(count(stone, 25) for stone in stones))
```
